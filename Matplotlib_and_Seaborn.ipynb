## Histogram
  - Represented as bins
  - Bins represents the data points in the range
  - plt.hist(ele,bins=count)

```python
import numpy as np
std1=np.random.randint(18,45,100)
std2=np.random.randint(15,40,100)
plt.hist(std1,bins=7)
#plt.xlabel('age category')
#plt.ylabel('no.of students')
plt.show()
```

```python
plt.hist(std2,bins=10,color='c',orientation="horizontal")
```

```python
std1
```

```python
std2
```

```python
vals=[i for i in range(10,50,5)]
plt.hist(std1,vals,histtype='bar',orientation='vertical',color='y')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals,rwidth=0.7,histtype='bar',orientation="vertical",color='mc')
plt.title('Multiple Histograms')
```

```python
x=[15,25,30,51,24]
plt.pie(x)
plt.show()
```

```python
#adding labels
f=['avg','good','exc','abv avg','moderate']
plt.pie(x,labels=f)
plt.show()
```

```python
#adding own colors
f=['avg','good','exc','abv avg','moderate']
plt.pie(x,labels=f,colors=['lightgreen','cyan','skyblue','pink','violet'],autopct='%.f%%')  #assigning % to regions
plt.title("Sort by Responses")
plt.show()
```

```python
e=[1,1,1,1,1]
f=['avg','good','exc','abv avg','moderate']
plt.pie(x,labels=f,colors=['lightgreen','cyan','skyblue','pink','violet'],autopct='%.f%%',shadow=True,startangle=180,explode=e)  #Excluding some regions
plt.title('Sort by Responses')
plt.show()
```

```python
# Stackplot
a=[9,4,3,2,4]
b=[12,4,5,15,21]
c=[7,12,21,17,11]
d=np.array([1,2,5,3,8])
plt.stackplot(a,b,c,d)
plt.show()
```

```python
f=[1,2,3,4,5]
s={'a':[5,4,3,2,1],'b':[1,2,3,8,9],'c':[7,6,5,6,9]}
plt.stackplot(f,s.values())
plt.show()
```

```python
#adding labels
plt.stackplot(f,s.values(),labels=s.keys())
plt.legend(loc='upper left')

plt.title('Stack Plot')
plt.show()
```
